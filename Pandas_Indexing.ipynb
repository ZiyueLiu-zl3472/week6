```python
import pandas as pd
```

```python
names = ['alan','bob','charlie']
ages = [20,30,40] #a list of numerics, not numpy
city = ['NY','NY','TX']
```

```python
mydata = list(zip(names,ages,city))
```

```python
mydata
```

```python
df = pd.DataFrame(data = mydata)
```

```python
df = pd.DataFrame(data = mydata, columns=['names','ages','city'])
```

```python
df = pd.DataFrame(data = mydata, columns=['names','ages','city'],index = ['one','two','three'])
```

```python
# Filter
df.loc[df['names']=='alan']
```

```python
# Select
df.loc[:,['ages']]
```

```python
# Arrange
df.sort_values('ages',ascending=0) #sort on column age from big to small
```

```python
# Mutate
df['FaveCity'] = df['city'] == 'NY' #create a new column by processing one or more other columns come up with a new value
#- want to know if fav city is NY - create a column FaveCity, test if city=NY, if true, return is true
```

```python
# Mutate
df['Younger'] = df['ages'] <= 30
```

```python
df
```

```python
# Groupby & Summarize
df.groupby('city')["ages"].mean() #group by city and want to know mean ages of each city
```
