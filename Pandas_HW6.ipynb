```python
import pandas as pd
```

```python
names = ['alan','bob','charlie']
ages = [20,30,40]
city = ['NY','NY','TX']
mydata = list(zip(names,ages,city))
df = pd.DataFrame(data = mydata,columns=['names','ages','city'])
```

```python
df
```

```python
df.loc[df['names']=='alan','city'] #iloc - look for number, loc-look for place -row or column
#look for row that has name alan, give me the column of city
```

```python
ind_alan = df['names'] == 'alan' #indicator variable
```

```python
df.loc[ind_alan] #give me all rows where indicator variable is true
```

```python
# Filter
df.loc[df['names']=='alan']
```

```python
# Select
df.loc[:,['ages','names']] #give me all rows, for columns only want ages and names
```

```python
df.sort_values('names',ascending=True)
```

```python
# Arrange
df.sort_values('ages',ascending=1)
```

```python
df['inNYC'] = df['city'] == "NY" #mutate
```

```python
df
```

```python
# Mutate
df['FaveCity'] = df['city'] == 'NY'
```

```python
df
```

```python
# Groupby & Summarize
df.groupby('city')["ages"].mean().sort_values(ascending=0)
```

```python
df.groupby('inNYC')["ages"].mean().sort_values(ascending = 0)
```
